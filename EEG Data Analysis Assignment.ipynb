```python
# Upgrade MNE-Python to ensure compatibility
!pip install --upgrade mne

# Import necessary modules
import mne
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Check MNE version to confirm availability of psd_array_multitaper function
print(mne.__version__)

# Import the psd_array_multitaper function from mne.time_frequency
from mne.time_frequency import psd_array_multitaper

# Define the directory where the EEG files are stored
eeg_dir = 'C:/Users/omkod/Downloads/eeg-during-mental-arithmetic-tasks-1.0.0/eeg-during-mental-arithmetic-tasks-1.0.0'

# Define the path to the subject information CSV file
subject_info_file = os.path.join(eeg_dir, 'subject-info.csv')

# Load subject information from CSV file
subject_info = pd.read_csv(subject_info_file)

# Display the first few rows of the dataframe
print(subject_info.head())

# Function to load EDF files
def load_edf(file_path):
    raw = mne.io.read_raw_edf(file_path, preload=True)
    return raw

# Define frequency bands
bands = {
    'Delta': (1, 4),
    'Theta': (4, 8),
    'Alpha': (8, 12),
    'Beta': (12, 30),
    'Gamma': (30, 100)
}

# Function to compute band-wise PSD
def compute_band_psd(raw, bands):
    data = raw.get_data()
    sfreq = raw.info['sfreq']
    psds, freqs = psd_array_multitaper(data, sfreq=sfreq, fmin=1, fmax=100, bandwidth=4, adaptive=True, normalization='full', verbose=0)
    psd_band = {}
    for band, (fmin, fmax) in bands.items():
        band_idx = np.logical_and(freqs >= fmin, freqs <= fmax)
        psd_band[band] = np.mean(psds[:, band_idx], axis=1)
    return psd_band

# List all files in the directory
files = os.listdir(eeg_dir)

# Filter and sort subject files
subject_files = sorted([f for f in files if f.endswith('.edf')])

# Initialize dictionaries to store PSDs for each subject
background_psds = {}
task_psds = {}

# Iterate through the subject files and process each subject
for i in range(0, len(subject_files), 2):
    background_file = os.path.join(eeg_dir, subject_files[i])
    task_file = os.path.join(eeg_dir, subject_files[i + 1])
    
    # Load the background and task EEG recordings
    background_raw = load_edf(background_file)
    task_raw = load_edf(task_file)
    
    # Compute band-wise PSD for background and task states
    background_psd = compute_band_psd(background_raw, bands)
    task_psd = compute_band_psd(task_raw, bands)
    
    # Store PSDs in dictionaries
    subject_id = subject_files[i].split('_')[0]
    background_psds[subject_id] = background_psd
    task_psds[subject_id] = task_psd

# Function to plot PSD comparison
def plot_psd_comparison(subject_id, background_psd, task_psd):
    plt.figure(figsize=(10, 6))
    width = 0.35  # the width of the bars
    ind = np.arange(len(bands))  # the x locations for the groups
    
    # Plot background PSD
    plt.bar(ind, [background_psd[band].mean() for band in bands.keys()], width, label='Background')
    
    # Plot task PSD
    plt.bar(ind + width, [task_psd[band].mean() for band in bands.keys()], width, label='Task')
    
    plt.xlabel('Frequency Band')
    plt.ylabel('Power Spectral Density (dB/Hz)')
    plt.title(f'Subject {subject_id}: PSD Comparison')
    plt.xticks(ind + width / 2, bands.keys())
    plt.legend(loc='best')
    plt.show()

# Example usage: Plot PSD comparison for the first subject in the list
if subject_files:
    example_subject_id = subject_files[0].split('_')[0]
    plot_psd_comparison(example_subject_id, background_psds[example_subject_id], task_psds[example_subject_id])
```

```python
# Step 3
# Import necessary modules for deep learning and data processing
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.constraints import max_norm

# Load and process the EEG data
# Note: Assuming `background_psds`, `task_psds`, and `subject_info` are already loaded

# Extract features and labels from the PSD data
features = []
labels = []

for subject_id in background_psds.keys():
    background_feature = [background_psds[subject_id][band].mean() for band in bands.keys()]
    task_feature = [task_psds[subject_id][band].mean() for band in bands.keys()]
    features.append(background_feature + task_feature)
    
    # Label based on the group information in the subject-info.csv
    label = subject_info[subject_info['Subject'] == subject_id]['Count quality'].values[0]
    labels.append(label)

features = np.array(features)
labels = np.array(labels)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the EEGNet model
def create_eegnet(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(16, (1, 4), padding='same', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.DepthwiseConv2D((1, 4), use_bias=False, depth_multiplier=2, depthwise_constraint=max_norm(1.)))
    model.add(layers.BatchNormalization())
    model.add(layers.Activation('elu'))
    model.add(layers.AveragePooling2D((1, 2)))
    model.add(layers.Dropout(0.5))
    
    model.add(layers.SeparableConv2D(16, (1, 4), padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.Activation('elu'))
    model.add(layers.AveragePooling2D((1, 2)))
    model.add(layers.Dropout(0.5))
    
    model.add(layers.Flatten())
    model.add(layers.Dense(2, activation='softmax'))
    
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Reshape the data to fit the EEGNet input shape
X_train_eegnet = X_train.reshape(X_train.shape[0], 1, X_train.shape[1], 1)
X_test_eegnet = X_test.reshape(X_test.shape[0], 1, X_test.shape[1], 1)

# Create and train the EEGNet model
eegnet_model = create_eegnet((1, X_train.shape[1], 1))
eegnet_model.fit(X_train_eegnet, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the EEGNet model
y_pred_eegnet = np.argmax(eegnet_model.predict(X_test_eegnet), axis=1)
print("EEGNet Model")
print("Accuracy:", accuracy_score(y_test, y_pred_eegnet))
print("Precision:", precision_score(y_test, y_pred_eegnet))
print("Recall:", recall_score(y_test, y_pred_eegnet))
print("F1 Score:", f1_score(y_test, y_pred_eegnet))

# Define the TSCeption model (simplified version for demonstration)
def create_tsception(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (1, 2), activation='relu', padding='same', input_shape=input_shape))
    model.add(layers.Conv2D(32, (1, 2), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((1, 2)))
    model.add(layers.Dropout(0.5))
    
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(2, activation='softmax'))
    
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Reshape the data to fit the TSCeption input shape
X_train_tsception = X_train.reshape(X_train.shape[0], 1, X_train.shape[1], 1)
X_test_tsception = X_test.reshape(X_test.shape[0], 1, X_test.shape[1], 1)

# Create and train the TSCeption model
tsception_model = create_tsception((1, X_train.shape[1], 1))
tsception_model.fit(X_train_tsception, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the TSCeption model
y_pred_tsception = np.argmax(tsception_model.predict(X_test_tsception), axis=1)
print("TSCeption Model")
print("Accuracy:", accuracy_score(y_test, y_pred_tsception))
print("Precision:", precision_score(y_test, y_pred_tsception))
print("Recall:", recall_score(y_test, y_pred_tsception))
print("F1 Score:", f1_score(y_test, y_pred_tsception))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.constraints import max_norm  # Import max_norm constraint

# Create EEGNet Model
def create_eegnet(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(16, (1, 4), padding='same', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.DepthwiseConv2D((1, 4), use_bias=False, depth_multiplier=2, depthwise_constraint=max_norm(1.)))
    model.add(layers.BatchNormalization())
    model.add(layers.Activation('elu'))
    model.add(layers.AveragePooling2D((1, 4)))
    model.add(layers.Dropout(0.25))
    
    # Adjust pooling and convolution parameters based on input shape
    if input_shape[1] >= 8:
        model.add(layers.SeparableConv2D(16, (1, 8), use_bias=False, padding='same'))
        model.add(layers.BatchNormalization())
        model.add(layers.Activation('elu'))
        model.add(layers.AveragePooling2D((1, input_shape[1] // 8)))  # Dynamic pooling size based on input shape
        model.add(layers.Dropout(0.25))
    else:
        model.add(layers.Flatten())
    
    model.add(layers.Flatten())
    model.add(layers.Dense(1, activation='sigmoid', kernel_constraint=max_norm(0.5)))
    
    return model

# Early Stopping Callback
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Reshape the data to fit the EEGNet input shape
X_train_eegnet = X_train.reshape(X_train.shape[0], 1, X_train.shape[1], 1)
X_test_eegnet = X_test.reshape(X_test.shape[0], 1, X_test.shape[1], 1)

# Compile and Train EEGNet Model
eegnet_model = create_eegnet((1, X_train.shape[1], 1))
eegnet_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = eegnet_model.fit(X_train_eegnet, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

```python
# Improved TSCeption Model
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

# Create TSCeption Model
def create_tsception(input_shape):
    model = models.Sequential()
    model.add(layers.Reshape((input_shape[0], input_shape[1], 1), input_shape=input_shape))
    model.add(layers.Conv2D(32, (10, 1), activation='relu'))
    model.add(layers.GlobalMaxPooling2D())
    model.add(layers.Dropout(0.5))
    
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(1, activation='sigmoid'))
    
    return model

# Early Stopping Callback
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Reshape the data to fit the TSCeption input shape
X_train_tsception = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_tsception = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Compile and Train TSCeption Model
tsception_model = create_tsception((X_train.shape[1], 1))
tsception_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = tsception_model.fit(X_train_tsception, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

```python
# Predictions from both models
y_pred_eegnet = eegnet_model.predict(X_test_eegnet)
y_pred_tsception = tsception_model.predict(X_test_tsception)

# Average the predictions for ensemble
y_pred_ensemble = (y_pred_eegnet + y_pred_tsception) / 2

# Convert predictions to binary
y_pred_ensemble_binary = (y_pred_ensemble > 0.5).astype(int)

# Evaluate EEGNet model
accuracy_eegnet = accuracy_score(y_test, (y_pred_eegnet > 0.5).astype(int))
precision_eegnet = precision_score(y_test, (y_pred_eegnet > 0.5).astype(int))
recall_eegnet = recall_score(y_test, (y_pred_eegnet > 0.5).astype(int))
f1_eegnet = f1_score(y_test, (y_pred_eegnet > 0.5).astype(int))

print(f'EEGNet Model - Accuracy: {accuracy_eegnet}, Precision: {precision_eegnet}, Recall: {recall_eegnet}, F1 Score: {f1_eegnet}')

# Evaluate TSCeption model
accuracy_tsception = accuracy_score(y_test, (y_pred_tsception > 0.5).astype(int))
precision_tsception = precision_score(y_test, (y_pred_tsception > 0.5).astype(int))
recall_tsception = recall_score(y_test, (y_pred_tsception > 0.5).astype(int))
f1_tsception = f1_score(y_test, (y_pred_tsception > 0.5).astype(int))

print(f'TSCeption Model - Accuracy: {accuracy_tsception}, Precision: {precision_tsception}, Recall: {recall_tsception}, F1 Score: {f1_tsception}')

# Evaluate the ensemble
accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble_binary)
precision_ensemble = precision_score(y_test, y_pred_ensemble_binary)
recall_ensemble = recall_score(y_test, y_pred_ensemble_binary)
f1_ensemble = f1_score(y_test, y_pred_ensemble_binary)

print(f'Ensemble Model - Accuracy: {accuracy_ensemble}, Precision: {precision_ensemble}, Recall: {recall_ensemble}, F1 Score: {f1_ensemble}')
```
